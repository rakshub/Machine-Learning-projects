```python
df=df.dropna()
```

```python
x=df.drop("label",axis=1)
```

```python
y=df["label"]
```

```python
print("X shape:",x.shape)
print("Y shape:",y.shape)
```

```python
import tensorflow as tf
tf.__version__
```

Tensorflow Libraries

```python
from tensorflow.keras .layers import Embedding #word2vec
from tensorflow.keras.preprocessing.sequence import pad_sequences #converting all sentence of equal length
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM , Dense
```

```python
voc_size=5000
```

```python
msg=x.copy()
```

```python
msg.reset_index(inplace=True)
```

```python
import nltk
import re #regular expression
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(msg)):
    review=re.sub('[^a-zA-Z]',' ',msg['title'][i]) # removing all special character expect alphabets
    review=review.lower()
    review=review.split()
    # if words are not present in english stop words stem it
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
ohe_rep=[one_hot(words,voc_size) for words in corpus]
ohe_rep
```

```python
sent_length=20
embed=pad_sequences(ohe_rep,maxlen=sent_length,padding="pre")
print(embed)
```

```python
embed_vector_feature=40
model=Sequential()
model.add(Embedding(voc_size,embed_vector_feature,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid')) #classification
model.compile(loss="binary_crossentropy",optimizer="adam",metrics=["accuracy"])
print(model.summary())
```

```python
x_final=np.array(embed)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.2,random_state=42)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=11,batch_size=80)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred=np.where(y_pred>0.5,1,0)
# 1 -it is a fake news
#0-it isn't a fake news
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
